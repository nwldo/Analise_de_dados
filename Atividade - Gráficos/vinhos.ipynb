```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../bases de dados/drinks.csv', sep=',')
```

```python
df.head(n=5)
```

### Análise dos países que consomem mais vinho

```python
# Ordena os dados pelo consumo de vinho em ordem decrescente
ordenar_dados = df.sort_values(by='wine_servings', ascending=False)

# Selecionar os 10 Países com maior consumo de vinho
top10_paises = ordenar_dados.head(10)

# gráfico de barras horizontais
plt.figure(figsize=(10, 6))
top10_paises.plot(x='country', y='wine_servings', kind='barh')
plt.title('Consumo de Vinho por País')
plt.xlabel('Consumo de Vinho (porções)')
plt.ylabel('País')
plt.show()
```

### Análise correlação entre o consumo de vinho e o consumo total de álcool puro

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['wine_servings'], df['total_litres_of_pure_alcohol'])
plt.title('Consumo de Vinho vs. Total de Litros de Álcool Puro')
plt.xlabel('Consumo de Vinho (servings)')
plt.ylabel('Total Litros de Álcool Puro')
plt.grid(True)
plt.show()
```

```python
df['proporcao_vinho'] = (df['wine_servings'] / df['total_litres_of_pure_alcohol']) * 100
df
```

```python
# Histograma
plt.hist(df['proporcao_vinho'], bins=20)
plt.xlabel('Proporção de Consumo de Vinho (%)')
plt.ylabel('Número de Países')
plt.title('Distribuição da Proporção de Consumo de Vinho')
plt.show()
```
