```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../bases de dados/drinks.csv', sep=',')
```

```python
df.head(n=5)
```

### Análise dos países que consomem mais vinho

```python
# filtro para obter as colunas country e wine_servings, e depois ordenado em ordem decrescente pelo consumo de vinho.
top_10_paises = df[['country', 'wine_servings']].sort_values(by='wine_servings', ascending=False).head(10)

# Criando o gráfico de barras horizontais
plt.figure(figsize=(10, 6))
plt.barh(top_10_paises['country'], top_10_paises['wine_servings'], color='darkred')
plt.title('Top 10 Países com Maior Consumo de Vinho')
plt.xlabel('Porções de Vinho por Pessoa por Ano')
plt.ylabel('País')
plt.gca().invert_yaxis()  # Inverter eixo Y para mostrar o maior valor no topo
plt.show()
```

```python
correlacao = df['wine_servings'].corr(df['total_litres_of_pure_alcohol'])

# Exibir a correlação calculada
print(f"Correlação entre consumo de vinho e consumo total de álcool: {correlacao:.2f}")
```

A correlação entre o consumo de vinho e o consumo total de álcool puro é de aproximadamente 0,67, indicando uma correlação positiva moderada. Isso significa que, em geral, países que consomem mais vinho também tendem a consumir mais álcool no total.

### Análise da Distribuição do Consumo de Vinho

```python
# Histograma
plt.hist(df['proporcao_vinho'], bins=20)
plt.xlabel('Proporção de Consumo de Vinho (%)')
plt.ylabel('Número de Países')
plt.title('Distribuição da Proporção de Consumo de Vinho')
plt.show()
```

O histograma mostra que a maioria dos países tem um consumo relativamente baixo de vinho, com alguns poucos países tendo um consumo muito elevado.
