```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# python -m pip install seaborn
import seaborn as sns
```

```python
#carregar o arquivo .csv: criar uma variável dataset que receberá a chamada de **pd.read_csv(arquivo)**
df = pd.read_csv('../bases de dados/carros.csv', sep=';')
```

```python
df.head(n=5)
```

### Realizar análise de ano de frabricação e o valor

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Ano'], df['Valor'], alpha=0.5)
plt.xlabel('Ano de Fabricação')
plt.ylabel('Valor do Veículo')
plt.title('Relação entre Ano de Fabricação e Valor')
plt.grid(True)
plt.show()
```

Com base nesses dados, pode se explorar Distribuição de Valores por Ano de Fabricação É possível observar uma tendência geral de aumento nos valores dos veículos ao longo dos anos, especialmente para os modelos mais recentes.

Outliers: Há alguns pontos fora da tendência geral, que podem representar veículos com características únicas ou negociações especiais.

### Distribuição do tipos de motores

```python
qtd_motor = df['Motor'].value_counts()

# Criando um gráfico de barras
sns.countplot(x='Motor', data=df)
plt.xlabel('Tipo de Motor')
plt.ylabel('Quantidade de Veiculos')
plt.title('Distribuição dos Tipos de Motor')
plt.xticks(rotation=45)
plt.show()
```

O gráfico mostra quais tipos de motor são mais comuns na base de dados(Motor 4.0 Turbo, Motor 1.0 8v, etc)

É possível comparar a popularidade de diferentes tipos de motor e identificar os mais comuns.
