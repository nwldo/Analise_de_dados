```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# python -m pip install seaborn
import seaborn as sns
```

```python
#carregar o arquivo .csv: criar uma variável dataset que receberá a chamada de **pd.read_csv(arquivo)**
df = pd.read_csv('../bases de dados/carros.csv', sep=';')
```

```python
df.head(n=10)
```

```python
#Retirando os valores NaN
df.fillna(0, inplace = False)
```

```python
df.Motor.value_counts().plot(kind='pie', autopct='%.2f%%')
plt.axis('equal');  # Para deixar o gráfico redondo
```

O gráfico mostra quais tipos de motor são mais comuns na base de dados(Motor V8, Motor Diesel, Motor 3.0 32v, etc)

É possível comparar a popularidade de diferentes tipos de motor e identificar os mais comuns.

```python
# Filtrar apenas os motores V8
df_v8 = df[df['Motor'] == 'Motor V8']

# Criar o gráfico de linha
sns.lineplot(x='Ano', y='Valor', hue='Motor', data=df_v8)
plt.title('Variação do Valor de Motores V8 ao Longo dos Anos')
plt.xlabel('Ano')
plt.ylabel('Valor')
plt.show()
```

Visualizar a evolução do valor do motore V8 indica que teve picos de valorização, mas teve momentos de desvalorização. Isso requer uma análise mais profunda.
