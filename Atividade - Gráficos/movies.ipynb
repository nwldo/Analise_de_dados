```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../bases de dados/movies.csv')
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Genre', y='Profitability', data=df)
plt.title('Lucratividade por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Lucratividade')
plt.show()
```

Os gêneros "Romance" e "Comedy" parecem ter uma mediana mais alta, sugerindo que, em média, filmes desses gêneros tendem a ser mais lucrativos.

```python
# Removendo o caractere '$' e convertendo a coluna para float
df['Worldwide Gross'] = df['Worldwide Gross'].str.replace('$', '').astype(float)

# Calculando a arrecadação global média por ano
arrecadacao_media = df.groupby('Year')['Worldwide Gross'].mean()

# Criando o gráfico de linha
plt.figure(figsize=(10, 6))
plt.plot(arrecadacao_media.index, arrecadacao_media.values)
plt.title('Arrecadação Global Média por Ano')
plt.xlabel('Ano')
plt.ylabel('Arrecadação Global Média')
plt.show()
```

A arrecadação global média dos filmes apresentou uma tendência de crescimento entre os anos de 2007 e 2008, atingindo um pico em 2008. No entanto, a partir de 2008, houve uma queda significativa na arrecadação média, atingindo o ponto mais baixo em 2010. Nos anos seguintes, houve uma leve recuperação, mas ainda abaixo do pico de 2008
