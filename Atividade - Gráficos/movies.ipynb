```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../bases de dados/movies.csv')
df.head()
```

```python
df.Genre.value_counts().plot(kind='pie', autopct='%.2f%%')
plt.axis('equal');  # Para deixar o gr√°fico redondo
```

Os g√™neros "Com√©dia", Romance e "Drama" s√£o mais populares, filmes desses g√™neros tendem a ser mais lucrativos.

```python
# Removendo o caractere '$' e convertendo a coluna para float
df['Worldwide Gross'] = df['Worldwide Gross'].str.replace('$', '').astype(float)

# Calculando a arrecada√ß√£o global m√©dia por ano
arrecadacao_media = df.groupby('Year')['Worldwide Gross'].mean()

# Criando o gr√°fico de linha
plt.figure(figsize=(10, 6))
plt.plot(arrecadacao_media.index, arrecadacao_media.values)
plt.title('Arrecada√ß√£o Global M√©dia por Ano')
plt.xlabel('Ano')
plt.ylabel('Arrecada√ß√£o Global M√©dia')
plt.show()
```

A arrecada√ß√£o global m√©dia dos filmes apresentou uma tend√™ncia de crescimento entre os anos de 2007 e 2008, atingindo um pico em 2008. No entanto, a partir de 2008, houve uma queda significativa na arrecada√ß√£o m√©dia, atingindo o ponto mais baixo em 2010. Nos anos seguintes, houve uma leve recupera√ß√£o, mas ainda abaixo do pico de 2008
