```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../bases de dados/drinks.csv', sep=',')
```

```python
df.head(n=5)
```

### Análise dos países consomem mais vinho

```python
plt.figure(figsize=(10, 6))
df.sort_values(by='wine_servings', ascending=False).plot(x='country', y='wine_servings', kind='barh')
plt.title('Consumo de Vinho por País')
plt.xlabel('Consumo de Vinho (porções)')
plt.ylabel('País')
plt.show()
```
